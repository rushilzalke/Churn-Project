```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Dell\OneDrive\Desktop\Churn\data\Telco-Customer-Churn.csv")
df.head()
```

```python
df.shape
df.info()
df.nunique()
df.isnull().sum()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors='coerce')
df.isnull().sum()
```

```python
sns.countplot(data=df, x="Churn")
plt.show()
```
